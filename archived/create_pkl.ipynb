```python
import pandas as pd
import numpy as np
from spacy.lang.en import English
import spacy
import glob
from pathlib import Path
import json
from tqdm import tqdm
```

```python
SILVER_FOLDER = '/Users/yisisang/Desktop/Character/imsdb_copy/imsdb_copy/silver'
read_files = glob.glob(str(Path(SILVER_FOLDER)) + "/*.txt")
```

```python
len(read_files)
```

```python
train_df = pd.DataFrame(columns=['label', 'title', 'sentence', 'book_id', 'section_id','sentence_id'])

for script_file in tqdm(read_files):
    with open(script_file, 'r') as fp:
        book_id = Path(script_file).stem
        lines = fp.readlines()
        rows_list = list()
        for sec_idx, line in enumerate(lines):
            sample = json.loads(line)
            #sample[2] = sample[2].strip('\n').replace('\n', ' ')
            #sents  = [str(sent) for sent in nlp(sample[2]).sents]
           #for sent_idx, sent in enumerate(sents):
            rows_list.append({
                'label':        sample[0],
                'title':        sample[1],
                'sentence':     sample[2],
                'book_id':      book_id,
                'section_id':   sec_idx,
                'sentence_id':  0,
            })
        train_df = train_df.append(rows_list)

output_file = "silver_imsdb_self_collected_add_space.pkl"
train_df.to_pickle(output_file)
print('[*] # of training samples: ', train_df.shape[0])
print('[*] Saved in <{}>'.format(output_file))
```
